```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Df= {'name' :['Ayobami', 'Isaac', 'Ebun', 'Segun', 'Michael', 'John',
              'Mojeed', 'Ayooluwa', 'Ezekiel', 'Elizabeth'], 
     'Age' :[23,24,21,18,30,25,28,24,22,27], 
     'Department':['Urban and Regional Planning',
                  'Estate', 'Surveying and Geo', 
                  'QTS', 'Architeture', 'Geology', 'Mathematics',
                  'Physics', 'Fine Art', 'Agric'], 
     'Level':[2,4,3,5,4,1,2,3,5,3]}
```

```python
Data = pd.DataFrame(Df)
```

```python
Data
```

```python
sns.barplot(data=Data, x='Age', y='name')
```

```python
sns.barplot(data=Data, x='name', y='Age')
```

```python
sns.kdeplot(data=Data, x='Level', y='Age')
```

```python

sns.catplot(data=Data, x ='name', y='Age',
           kind='bar', height=4, aspect=3)
```

```python
fig= plt.subplots(figsize=(20,5))
sns.lineplot(data=Data, x ='name', y='Age')
```

```python
fig= plt.subplots(figsize=(10,10))
sns.histplot(data=Data, x ='Age',  bins=5)
```

```python
fig= plt.subplots(figsize=(10,10))
sns.histplot(data=Data, x ='Level',  bins=5)
```
